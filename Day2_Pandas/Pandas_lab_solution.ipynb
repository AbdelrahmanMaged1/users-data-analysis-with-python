`.mode()` method in pandas returns a Series, not a single value. 
Even if there is only one most frequent date, pandas provides it as a list-like object, which `fillna()` doesn't know how to map to a single missing cell.
So, to select the first element of that Series I should use [0].

```python
# •Fill missing Score Date values with the most common date (mode).
print(df['Score Date'])

print(df['Score Date'].fillna( df['Score Date'].mode()[0] ) )
```

#### •Find the student(s) with the highest Grade and show their Name and Grade.
##### This is explanation and clarification for me to remember, NOT CHEATING.

**1. Using `nlargest()`** 

Is often the most readable method for getting the "top" items. It is specifically designed to return the top `n` rows.

with option/parameter `keep="all"` ensures that if there is a tie for first place, both/all students are returned.

---
**2. Using `rank()`**

`rank()` assigns a numerical rank to every student. I can then filter by `rank == 1`.

---
### Which one should we choose?

| Method | Best For... |
| --- | --- |
| **Boolean Filtering** (`df['Grade'] == max()`) | Simple, direct, and very fast for single conditions. |
| **`nlargest()`** | Getting the top $N$ results (e.g., "Top 3 students") while handling ties easily. |
| **`rank()`** | Complex scenarios where you need to know the relative position of everyone in the class. |


```python
# •Find the student(s) with the highest Grade and show their Name and Grade.Lab
# using ---------------> boolean condition
# df[ condition ] [ list of wanted columns ]
df[ df['Grade'] == df['Grade'].max() ] [['Name', 'Grade']]

# or using .loc()
df.loc[df['Grade'] == df['Grade'].max(), ['Name', 'Grade']]


# or using ------------> .nlargest() <------------                  
# 'keep="all"' ensures that if there is a tie for first place, both/all students are returned.
top_students= df.nlargest(1, 'Grade', keep='all')[['Name', 'Grade']]
top_students

# or using ------------> .rank() <------------                  
# 'min' method ensures that tied students get the same top rank
# Filter for the students ranked 1

df['Rank'] = df['Grade'].rank(method='min', ascending=False)

top_students = df[df['Rank'] == 1][['Name', 'Grade']]
top_students
```

```python
# •After filling missing grades, recalculate the average Grade per Subject.

df['Grade'] = df['Grade'].fillna(df['Grade'].mean())

avg_per_subject = df.groupby('Subject')['Grade'].mean()

avg_per_subject
```

```python
# •Create a new column City_Avg that shows the average grade of each student's city (using transform).
# transform() performs the calculation for each group and broadcasts the result back to the original index size, 
# allowing you to easily assign it as a new column.
df['City_Avg'] = df.groupby('City')['Grade'].transform('mean')

df
```
