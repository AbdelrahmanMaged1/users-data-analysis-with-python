```python
a = np.array([1, 2, 3, 4])          # shape (4,) -> behaves like (1,4)
b = np.array([[10], [20], [30]])    # shape (3,1)  


print(a+b)

# When performing a + b, NumPy applies broadcasting:
# a is treated as shape (1, 4).
# b has shape (3, 1).

# Broadcasting rules expand dimensions where size = 1:
# a -> repeated along rows to shape (3,4)
# b -> repeated along columns to shape (3,4)

# Then element-wise addition is performed.
# Final result shape: (3,4)
```

```python
data = np.array([150, 220, 90, 300, 180, 45])
print(f'{data}\n')

norm_data = (data - data.min()) / (data.max() - data.min())
print(norm_data)
```

8. Standardize (z-score) the same array from question 7.

```python
# 8. Standardize (z-score) the same array from question 7.

# Standardization (Z-score Scaling)
# ◦ Centers data to mean = 0 and standard deviation = 1

data = np.array([150, 220, 90, 300, 180, 45])

stand = (data - data.mean()) / data.std()
print(stand)
```

```python
mat = np.array([
    [4,  8,  1, 12],
    [7,  3,  9,  5],
    [11, 2,  6, 10],
    [15, 0, 14, 13]
])

# a
row_mean = np.mean(mat, axis=1, keepdims=True)
print(f'row mean\n{row_mean}\n')

# b
col_sum = np.sum(mat, axis=0, keepdims=True)
print(f'col_sum\n{col_sum}\n')

# c
max_val = np.max(mat)
print(f'max value: {max_val}')

# np.argmax gives the flat index; 
# unravel_index converts it to (row, col)
max_position = np.unravel_index(np.argmax(mat), mat.shape)
print(f'max_position: {max_position}')
```

```python
ages = np.array([23, 45, 19, 34, 28, 51, 17, 39])
names = np.array(["Ali", "Sara", "Omar", "Lina", "Khaled", "Nour", "Yara", "Hassan"])

# first we get the wanted range of ages as mask (boolean array)
mask = (25 <= ages) & (ages <= 40)

# then we apply the mask (wanted ages) on the names array to get only the "True" names
wanted_names = names[mask]
print(wanted_names)
```

```python
arr = np.arange(1, 21)
print(f'array:\n{arr}\n')

# shuffle operates in place
np.random.shuffle(arr)
print(f'shuffled array:\n{arr}\n')

# to keep the original as it is and return a new shuffled copy we use permutation
shuffled = np.random.permutation(arr)
print(f'shuffled array:\n{shuffled}')
```
